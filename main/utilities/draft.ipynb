```python
import fitz  # PyMuPDF
from docx import Document
```

```python
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = []

    for page in doc:
        blocks = page.get_text("blocks")  # layout-aware text blocks
        blocks = sorted(blocks, key=lambda b: (b[1], b[0]))  # sort by vertical then horizontal pos

        left_col = []
        right_col = []

        # Automatically detect median x to split columns
        widths = [b[0] for b in blocks]
        median_x = page.rect.width / 2

        for block in blocks:
            x0, y0, x1, y1, text, *_ = block
            if x0 < median_x:
                left_col.append((y0, text))
            else:
                right_col.append((y0, text))

        left_col.sort()
        right_col.sort()
        page_text = [t for _, t in left_col + right_col]
        full_text.append("\n".join(page_text))

    return "\n\n".join(full_text)

def save_text_to_word(text, output_path):
    doc = Document()
    for paragraph in text.split("\n\n"):
        doc.add_paragraph(paragraph.strip())
    doc.save(output_path)


```

```python
# Usage
pdf_path = r"C:\Users\User_1\Desktop\Translation_samples\other_docs\article_1_3p.pdf"
output_path = r"C:\Users\User_1\Desktop\Translation_samples\other_docs\doc_article_1_3p.docx"

text = extract_text_from_pdf(pdf_path)
save_text_to_word(text, output_path)

print("✅ PDF text extracted and saved to Word.")
```
