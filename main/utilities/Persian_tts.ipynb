```python
from TTS.utils.synthesizer import Synthesizer
import unicodedata

# Set local paths
model_path = r"D:\Projects\Offline_translator-webapp\tts_models\checkpoint_88000.pth"
config_path = r"D:\Projects\Offline_translator-webapp\tts_models\config.json"
output_path = r"C:\Users\User_1\Desktop\tts_o2.wav"

# Initialize synthesizer
synthesizer = Synthesizer(
    tts_checkpoint=model_path,
    tts_config_path=config_path,
    vocoder_checkpoint=None,
    vocoder_config=None,
    use_cuda=False  # True if you want to use GPU
)

# Persian text
text = "زندگی فقط یک بار است از آن به خوبی استفاده کن"
text = unicodedata.normalize("NFC", text)


# Generate audio waveform
wavs = synthesizer.tts(text)

# Save to file
synthesizer.save_wav(wavs, output_path)

print("✅ Audio saved at:", output_path)
```

```python
import asyncio
from persianttsfarsi import PersianTTS

output_path = r"C:\Users\User_1\Desktop\tts_o3.mp3"

tts = PersianTTS()

# تابع کمکی برای اجرای تابع async
async def run_tts():
    await tts._synthesize_ssml(
        text="سلام دنیا! این یک مثال از تبدیل متن به گفتار فارسی است.",
        output_file=output_path
    )

# اجرا
await run_tts()
print("✅ فایل صوتی تولید شد:", output_path)
```

```python
import pyttsx3

engine = pyttsx3.init()

# Check available voices
voices = engine.getProperty('voices')
for voice in voices:
    print(f"Voice: {voice.id}, Language: {voice.languages}")

# Set Persian voice (if available)
for voice in voices:
    if 'persian' in voice.id.lower():  # or check 'fa' language in voice.languages
        engine.setProperty('voice', voice.id)
        break

engine.say("سلام، چطور می توانم به شما کمک کنم؟")
engine.runAndWait()
```
