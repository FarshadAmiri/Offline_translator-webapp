```python
import os ; os.chdir(r"D:\Projects\Offline_translator-webapp\main\utilities")
from docx import Document
from docx.shared import RGBColor
from docx.shared import Pt
from docx.enum.style import WD_STYLE_TYPE
from docx.oxml import OxmlElement
from docx.oxml.ns import qn
from argos import translate

data_folder = r"C:\Users\User_1\Desktop\translation_tmp\my_module"
```

```python
def simplify_formatting(doc):
    for paragraph in doc.paragraphs:
        for run in paragraph.runs:
            run.font.bold = True
            run.font.italic = True
            run.font.underline = True
            run.font.color.rgb = None
    
    for paragraph in doc.paragraphs:
        for run in paragraph.runs:
            run.font.bold = False
            run.font.italic = False
            run.font.underline = False
            run.font.color.rgb = None
    
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    for run in paragraph.runs:
                        run.font.bold = False
                        run.font.italic = False
                        run.font.underline = False
                        run.font.color.rgb = None

    temp_simplified_doc_path = r'C:\Users\User_1\Desktop\temp_simplified_document.docx'
    doc.save(temp_simplified_doc_path)


def set_rtl(paragraph):
    # Set the direction of text to RTL (right-to-left)
    p_pr = paragraph._element.get_or_add_pPr()
    bidi = OxmlElement('w:bidi')
    bidi.set(qn('w:val'), 'true')
    p_pr.append(bidi)


def translate_document(input_path, output_path, source_lang, target_lang):
    doc = Document(input_path)
    simplify_formatting(doc)
    temp_simplified_doc_path = os.path.join(data_folder, "tmp_simplified_document.docx")
    doc.save(temp_simplified_doc_path)
    doc = Document(temp_simplified_doc_path)

    mystyle = doc.styles.add_style('mystyle', WD_STYLE_TYPE.CHARACTER)
    
    # Iterate over paragraphs and translate text
    for paragraph in doc.paragraphs:
        for run in paragraph.runs:
            translated_text = translate(run.text, source_lang, target_lang)
            run.style = mystyle
            run.text = translated_text
            run.font.size = Pt(run.font.size.pt) if run.font.size else None
            run.font.bold = run.font.bold
            run.font.italic = run.font.italic
            run.font.rtl = True
        set_rtl(paragraph)
    
    # Iterate over tables and translate text
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    for run in paragraph.runs:
                        translated_text = translate(run.text, source_lang, target_lang)
                        run.text = translated_text
                        run.font.size = Pt(run.font.size.pt) if run.font.size else None
                        run.font.bold = run.font.bold
                        run.font.italic = run.font.italic
                    set_rtl(paragraph)

    # Save the translated document
    doc.save(output_path)

# Example usage
input_docx_path =  os.path.join(data_folder, "source.docx")
output_docx_path = os.path.join(data_folder, "target.docx")
source_lang = 'en'  # Target language code, e.g., 'es' for Spanish
target_lang = 'fa'  # Target language code, e.g., 'es' for Spanish

translate_document(input_docx_path, output_docx_path, source_lang, target_lang)
```

```python
from docx import Document
from docx.shared import RGBColor

def simplify_formatting(doc):
    # Iterate through paragraphs and their runs
    for paragraph in doc.paragraphs:
        for run in paragraph.runs:
            run.font.bold = False
            run.font.italic = False
            run.font.underline = False
            run.font.color.rgb = RGBColor(0, 0, 0)  # Set color to black
            run.clear()  # Clear any hyperlink formatting

    # Iterate through tables and their cells
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for paragraph in cell.paragraphs:
                    for run in paragraph.runs:
                        run.font.bold = False
                        run.font.italic = False
                        run.font.underline = False
                        run.font.color.rgb = RGBColor(0, 0, 0)  # Set color to black
                        run.clear()  # Clear any hyperlink formatting

# Example usage
input_docx_path = r"C:\Users\User_1\Desktop\source.docx"
temp_simplified_doc_path = r'C:\Users\User_1\Desktop\temp_simplified_document.docx'


# Load the document
doc = Document(input_docx_path)

# Simplify formatting
simplify_formatting(doc)

# Save the modified document
doc.save(temp_simplified_doc_path)

print(f"Simplified document saved to {temp_simplified_doc_path}")
```

```python
from docx import Document
from docx.shared import RGBColor
from docx.oxml.ns import qn
import io

def convert_to_plain_text_with_images(input_docx_path, output_docx_path):
    # Load the input document
    doc = Document(input_docx_path)
    
    # Create a new Document for the output
    plain_doc = Document()

    def copy_images_from_paragraph(paragraph, new_paragraph):
        # Copy images from the original paragraph to the new paragraph
        for rel in paragraph.part.rels.values():
            if "image" in rel.target_ref:
                img_part = rel.target_part
                img_stream = img_part.blob
                try:
                    # Add image to the new paragraph
                    new_run = new_paragraph.add_run()
                    new_run.add_picture(io.BytesIO(img_stream))
                except Exception as e:
                    print(f"Failed to add image: {e}")

    def copy_images_from_table_cell(cell, new_cell):
        # Copy images from the original cell to the new cell
        for rel in cell.part.rels.values():
            if "image" in rel.target_ref:
                img_part = rel.target_part
                img_stream = img_part.blob
                try:
                    # Add image to the new cell
                    new_paragraph = new_cell.add_paragraph()
                    new_run = new_paragraph.add_run()
                    new_run.add_picture(io.BytesIO(img_stream))
                except Exception as e:
                    print(f"Failed to add image: {e}")

    # Copy paragraphs with plain text and retain images
    for paragraph in doc.paragraphs:
        new_paragraph = plain_doc.add_paragraph()
        
        for run in paragraph.runs:
            new_run = new_paragraph.add_run(run.text)
            new_run.bold = False
            new_run.italic = False
            new_run.underline = False
            new_run.font.color.rgb = RGBColor(0, 0, 0)  # Set color to black
        
        # Copy images if any
        copy_images_from_paragraph(paragraph, new_paragraph)
    
    # Copy tables with plain text and retain images
    for table in doc.tables:
        # Add a new table to the plain document with the same number of columns
        plain_table = plain_doc.add_table(rows=0, cols=len(table.columns))
        
        for row in table.rows:
            plain_row = plain_table.add_row().cells
            for idx, cell in enumerate(row.cells):
                # Set cell text
                plain_row[idx].text = cell.text
                # Copy images in the cell
                copy_images_from_table_cell(cell, plain_row[idx])

    # Save the new document as plain text with images preserved
    plain_doc.save(output_docx_path)


# Example usage
input_docx_path = r"C:\Users\User_1\Desktop\source.docx"
temp_simplified_doc_path = r'C:\Users\User_1\Desktop\temp_simplified_document.docx'

convert_to_plain_text_with_images(input_docx_path, temp_simplified_doc_path)
```
